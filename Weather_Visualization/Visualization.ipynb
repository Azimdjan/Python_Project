```python
import matplotlib.pyplot as plt
import pandas as pd

dataSet = pd.read_csv('../Data/KCLT.csv')
date = dataSet['date'].head(10)
actual_mean_temp = dataSet['actual_mean_temp'].head(10)

labels = date
sizes = actual_mean_temp

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
ax1.set_title("2014 USA's average temperature for 10 days in percentage")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

dataSet = pd.read_csv('../Data/KCLT.csv')


data = dataSet['record_precipitation'].tail(5)
ingredients = dataSet['date'].tail(5)


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Date",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("2015 USA's average precipitation for 5 days in percetange with value.")

plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)
dataSet = pd.read_csv('../Data/KCLT.csv')

plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
date = dataSet['date'].head(15)
y_pos = np.arange(len(date))
performance = dataSet['record_max_temp'].head(15)

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(date, color='red')
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel("Temperature 'F")
ax.set_title("2014 USA's record max temperatures for 15 days", color="#575353")

plt.show()
```
